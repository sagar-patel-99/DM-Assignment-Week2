```python
from datetime import date
#call dataset from drive
import pandas as pd
data =pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/company_revenue.csv")
data['Date']=pd.to_datetime(data['Date'])
data.head()
```

```python
from pycaret.time_series import setup

s = setup(data, target='Revenue', session_id=123)
```

```python
from pycaret.time_series.forecasting.functional import compare_models
best_model = compare_models(sort='MAE')  # sorts by Mean Absolute Error
```

```python
from pycaret.time_series.forecasting.functional import create_model
model = create_model('naive')
```

```python
from pycaret.time_series.forecasting.functional import tune_model


tuned_model = tune_model(model)
```

```python
from pycaret.time_series.forecasting.functional import plot_model


plot_model(tuned_model)
```

```python
from pycaret.time_series.forecasting.functional import predict_model


predictions = predict_model(tuned_model)
```

```python
from pycaret.time_series.forecasting.functional import finalize_model


final_model = finalize_model(tuned_model)
```
